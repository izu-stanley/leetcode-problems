```python
from collections import Counter


def checker(arr):
    counts = Counter(arr).values()
    if sum(counts) != len(counts):
        return False
    else:
        return True
def holder_func(x):
    if not checker(x):  return False

def helper(arr):
    holder = []
    holder1 = []
    holder2 = []
    holder3 = []
    holder4 = []
    holder5 = []
    holder6 = []
    holder7 = []
    holder8 = []
    holder9 = []
    box1 = []
    box2 = []
    box3 = []
    box4 = []
    box5 = []
    box6 = []
    box7 = []
    box8 = []
    box9 = []
    for x,y in enumerate(arr):

        #Checks Rows
        if not checker(holder): return False 
        holder = []
        for i,j in enumerate(y):

            if j == '.':
                continue
            else:
                holder.append(j)
        if x == 8:
            if not checker(holder): return False 


        #Checks Columns
        if y[0] == '.':
            pass
        else:
            holder1.append(y[0])


        if y[1] == '.':
            pass
        else:
            holder2.append(y[1])

        switch = 2
        if y[2] == '.':
            pass
        else:
            holder3.append(y[2])

        switch = 3
        if y[3] == '.':
            pass
        else:
            holder4.append(y[3])
        switch = 4
        if y[switch] == '.':
            pass
        else:
            holder5.append(y[switch])
        switch = 5
        if y[switch] == '.':
            pass
        else:
            holder6.append(y[switch])
        switch = 6
        if y[switch] == '.':
            pass
        else:
            holder7.append(y[switch])
        switch = 7
        if y[switch] == '.':
            pass
        else:
            holder8.append(y[switch])
        switch = 8
        if y[switch] == '.':
            pass
        else:
            holder9.append(y[switch])

        all_holders = [holder1,holder2,holder3,holder4,holder5,
                       holder6,holder7,holder8,holder9]  
        results = [holder_func(x) for x in all_holders]

        if False in results: return False


        #Checks Boxes
        for i,j in enumerate(y):
            if i >= 0 and i < 3 and x >= 0 and x < 3:
                if j == '.':
                    continue
                else:
                    box1.append(j)

        if not checker(box1): return False 


        for i,j in enumerate(y):
            if i >= 0 and i < 3 and x >= 3 and x < 6:
                if j == '.':
                    continue
                else:
                    box2.append(j)

        if not checker(box2): return False 


        for i,j in enumerate(y):
            if i >= 0 and i < 3 and x >= 6 and x < 9:
                if j == '.':
                    continue
                else:
                    box3.append(j)

        if not checker(box3): return False 

        for i,j in enumerate(y):
            if i >= 3 and i < 6 and x >= 0 and x < 3:
                if j == '.':
                    continue
                else:
                    box4.append(j)

        if not checker(box4): return False 


        for i,j in enumerate(y):
            if i >= 3 and i < 6 and x >= 3 and x < 6:
                if j == '.':
                    continue
                else:
                    box5.append(j)

        if not checker(box5): return False 


        for i,j in enumerate(y):
            if i >= 3 and i < 6 and x >= 6 and x < 9:
                if j == '.':
                    continue
                else:
                    box6.append(j)

        if not checker(box6): return False 


        for i,j in enumerate(y):
            if i >= 6 and i < 9 and x >= 0 and x < 3:
                if j == '.':
                    continue
                else:
                    box7.append(j)

        if not checker(box7): return False 

        for i,j in enumerate(y):
            if i >= 6 and i < 9 and x >= 3 and x < 6:
                if j == '.':
                    continue
                else:
                    box8.append(j)

        if not checker(box8): return False


        for i,j in enumerate(y):
            if i >= 6 and i < 9 and x >= 6 and x < 9:
                if j == '.':
                    continue
                else:
                    box9.append(j)

        if not checker(box9): return False 

    return True

```
